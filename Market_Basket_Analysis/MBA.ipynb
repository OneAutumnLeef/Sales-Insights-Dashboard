```python
import pandas as pd
from IPython.display import display

# 1. Load the data
df = pd.read_csv('MBA_Data.csv')
df = df.drop(columns=['Timestamp'])

# 2. Add TransactionID if missing
if 'TransactionID' not in df.columns:
    df.insert(0, 'TransactionID', range(1, len(df) + 1))

# 3. One‐hot encode items
item_list = [
    'Milk','Bread','Eggs','Cheese','Rice','Pasta/Noodles',
    'Shampoo','Detergent','Chips','Chocolate','Juice',
    'Vegetables','Fruits'
]
items_col = [c for c in df.columns if 'Items Purchased' in c][0]

def has_item(cell, item):
    if pd.isna(cell):
        return 0
    return int(item in [i.strip() for i in cell.split(',')])

for item in item_list:
    df[item] = df[items_col].apply(lambda x: has_item(x, item))

# 4. Correct segmentation of City into one‐hots
segment_col = 'City'
segment_map = {
    'Rural':     'Rural (Villagers/Small Towns)',
    'Urban':     'Urban (City/Metro)',
    'Suburban':  'Suburban (Outskirts)'
}
for seg_name, seg_val in segment_map.items():
    df[seg_name] = (df[segment_col] == seg_val).astype(int)

# 5. Build final basket dataframe
basket_df = df[['TransactionID'] + list(segment_map.keys()) + item_list]

# 6. Save & display
basket_df.to_csv('MBA_Basket_OneHot_Segmented.csv', index=False)
basket_df.to_excel('MBA_Basket_OneHot_Segmented.xlsx', index=False)
print(basket_df.head())
display(basket_df)
```

```python
# New cell: Frequent pattern mining and association rules
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from IPython.display import display

# Load one-hot encoded basket
df_onehot = pd.read_csv('MBA_Basket_OneHot_Segmented.csv')

# Run Apriori (e.g. min_support=5%)
freq_itemsets = apriori(
    df_onehot.drop(columns=['TransactionID']),
    min_support=0.05,
    use_colnames=True
)

# Generate rules (e.g. lift ≥ 1.2)
rules = association_rules(
    freq_itemsets,
    metric='lift',
    min_threshold=1.2
)

# Display top patterns and rules
display(freq_itemsets.sort_values('support', ascending=False).head(10))
display(rules.sort_values('lift', ascending=False).head(10))

# Save rules for later review
rules.to_csv('association_rules.csv', index=False)
print("Saved to association_rules.csv")
```

```python
import pandas as pd
# import ace_tools as tools

# Load the data
path = r'C:\Users\deraj\Hall of Codes\BA\Market_Basket_Analysis\MBA_Data.xlsx'
df = pd.read_excel(path)

# Define the metadata columns to exclude from items
meta_cols = [
    'Timestamp',
    'Name (Please use the same name if you\'re filling it multiple times)',
    'Age Group',
    'Gender',
    'When do you generally buy Grocery?',
    'Total Spend (On Average) Enter a number like 1000/5000 (no commas or words)',
    'Loyalty ( Do you keep going to the same place or change frequently? )',
    'City'
]

# Identify item columns by exclusion
item_cols = [col for col in df.columns if col not in meta_cols]

# Create TransactionID
df['TransactionID'] = df.index + 1

# Combine selected items into a list per transaction (non-null entries indicate selection)
df['Items'] = df[item_cols].apply(lambda row: [col for col in item_cols if pd.notna(row[col])], axis=1)

# Prepare the transactional dataframe
transactions_df = df[['TransactionID', 'Items'] + meta_cols]

# Display a sample (replacing tools.display_dataframe_to_user with standard pandas display)
print("Prepared Transaction Data Sample")
display(transactions_df.head())

transactions_df.to_csv('MBA_transactions.csv', index=False)
print("Exported transactions to MBA_transactions.csv")
```
